### Pre-processing of the data

 1. Handling null values
 2. Un-nesting the columns


#### 1.Handling null values

```python
df.isnull().sum()
```

```python
df.director.fillna('Unknown Director', inplace=True)
df.cast.fillna('Unknown Cast', inplace=True)
df.country.fillna('Unknown Country', inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=["date_added", "rating",'duration'], inplace=True)
df.isnull().sum()
```

```python
df.shape
```

```python
df.head(5)
```

```python
df.head()
```

#### 2.Unnesting the columns 
(cast, director, country,listed_in)

```python
df1= df[[ 'title','cast']]
```

```python
df1= df[[ 'title','cast']]
```

```python
def split_my_str(inp):
    return str(inp).split(', ')
```

```python
df1['cast'] = df1.cast.apply(split_my_str)
```

```python
df_cast= df1.explode('cast')
df_cast.head()
```

```python
df2 = df[[ 'title','director']]
df2['director'] = df2.director.apply(split_my_str)
```

```python
df2.head()
```

```python
df_director= df2.explode('director')
df_director.head()
```

```python
df3 = df[['title','country']]
df3.head()
```

```python
df3['country'] = df3.country.apply(split_my_str)
```

```python
df_country= df3.explode('country')
df_country.head()
```

```python
df4 = df[['title','listed_in']]
df4.head()
```

```python
df4['listed_in'] = df4.listed_in.apply(split_my_str)
```

```python
df_listed_in = df4.explode('listed_in')
df_listed_in.head()
```

#### joining all un-nested columns

```python
join1 = df_cast.merge(df_director,left_on='title',right_on='title')
```

```python
join2 = join1.merge(df_country,left_on='title',right_on='title')
```

```python
join3 = join2.merge(df_listed_in,left_on='title',right_on='title')
join3.head(5)
```

```python
df_partial = df[[ 'show_id','type','title','date_added','release_year','rating','duration','description']]
```

#### joining all the un-nested columns with the original dataframe creating 'df_final' dataframe

```python
df_final = join3.merge(df_partial,left_on='title',right_on='title')
df_final.head()
```

```python
df_final.isnull().sum()
```

```python
df_final['date_added_final'] = pd.to_datetime(df['date_added'],format='mixed')
df_final.head()
```

```python
df_final.info()
```

```python
df_final.shape
```

```python
movies= df_final[df_final['type'] == 'Movie']
movies.head()
movies.info()
```
